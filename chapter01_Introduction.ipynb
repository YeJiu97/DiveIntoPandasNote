## 1.1 什么是Pandas

**Python与Pandas**

Python语言：

>Python由荷兰人吉多·范罗苏姆（Guido van Rossum）创造，第一版发布于1991年。

Python语言可以进行：

1. Web开发
2. 网络爬虫
3. 计算与数据分析
4. GUI开发
5. 人工智能
6. 游戏开发
7. 图形图像

等等。

Pandas的基本功能：

- 从Excel、CSV、网页、SQL、剪贴板等文件或工具中读取数据；
- 合并多个文件或者电子表格中的数据，将数据拆分为独立文件；
- 数据清洗，如去重、处理缺失值、填充默认值、补全格式、处理极端值等；
- 建立高效的索引；
- 支持大体量数据；
- 按一定业务逻辑插入计算后的列、删除列；
- 灵活方便的数据查询、筛选；
- 分组聚合数据，可独立指定分组后的各字段计算方式；
- 数据的转置，如行转列、列转行变更处理；
- 连接数据库，直接用SQL查询数据并进行处理；
- 对时序数据进行分组采样，如按季、按月、按工作小时，也可以自定义周期，如工作日；
- 窗口计算，移动窗口统计、日期移动等；
- 灵活的可视化图表输出，支持所有的统计图形；
- 为数据表格增加展示样式，提高数据识别效率。

但是实际上name才是这个df的索引，所以我们需要对其进行修改：

````python
df.set_index('name', inplace = True)  # 建立索引并且生效
````

````python
df
````

````python
df.head()
````

````python
df['Q1']  # 单独将Q1提取出来
````

````python
df[['team', 'Q1']]  # 将team和Q1提取出来
````

````python
df[df.index == 'Liver']  # 指定姓名，将其所在行提取出来
````

````python
df[0:3]  # 提取前3行
````

````python
df[0:20:4]  # 在前20个中每4个选取1个
````

````python
df.loc['Ben', 'Q1':'Q4']  # 将ben的四个季度成绩都提取出来
````

````python
df.loc['Eorge':'Alexander', 'team':'Q4']
````

````python
df[df.Q1 > 80]  # 将Q1中所有大于80的都提取出来
````

````python
df[df.team == 'C']  # 所有team为‘C’的情况
````

````python
df.sort_values(by = 'Q1')  # 按Q1列数据升序排列
````

````python
df.groupby('team').sum()  # 按团队分组对应列相加
````

````python
df.groupby('team').mean()  # 按团队分组对应列相加
````

````python
df.groupby('team').agg({'Q1':sum,
                       'Q2':'count',
                       'Q3':'mean',
                       'Q4':max})
````
